Nessa questão será analisada uma base de dados de cereais utilizando listas de listas.


```python
import csv
# Faz a leitura do arquivo .csv e gera uma lista
with open('cereal.csv', newline='') as f:
    reader = csv.reader(f)
    data = list(reader)
```

```python
def get_nr_coluna(dados, nome_coluna):
  """Captura o índice da coluna, dado o nome informado."""

  for col in range(len(dados[0])):
    if(dados[0][col].startswith(nome_coluna)):
      return col
```

```python
def replacer(data, original, replaced):
  """
  Substitui na coluna MFR as siglas existentes em uma lista (original) pelos nomes correspondentes existentes em outra lista (replaced).
  Parâmetros: matriz com listas dos dados de cereais, lista com siglas a serem substituídas, lista com nomes de cereais correspondentes.
  """

  for row in data:
    for posicao, col in enumerate(row):
      if posicao == get_nr_coluna(data, "mfr"):
        for item, value in enumerate(original):
          if col == value:
            row[posicao] = replaced[item]

  return data
```

```python
original = ['A', 'G', 'K', 'N', 'P', 'Q', 'R', 'H', 'C']
replaced = ['AHFP', 'GM', 'Kelloggs', 'Nabisco', 'Post', 'Quaker', 'Purina', 'Hot', 'Cold']
```

Aplica a função replacer, verificando se os dados pertencem à coluna MFR e caso positivo, substitui as siglas existentes em uma lista pelos nomes do cereal correspondente (em outra lista).


```python
dados = replacer(data, original, replaced)
dados
```

```python
def replace_semicolon(data):
  """Verifica os dados pertencentes à coluna RATING e substitui as strings ';' por ''. """

  for row in data:
    for posicao, col in enumerate(row):
      if posicao == get_nr_coluna(data, "rating"):
        row[posicao] = (row[posicao]).replace(";", "")

  return data
```

Aplica a função replace_semicolon, que verifica os dados pertencentes à coluna RATING e removendo as strings ';;'.

```python
dados_novos = replace_semicolon(dados)
```

```python
def count_ratio_hot_cold(data):
  """
  Faz a contagem de valores Cold (C) e quantitativo total (Hot e Cold).
  Calcula a razão existente entre Cold/Registros totais, assim como Hot/Registros totais (excluindo a linha referente ao nome dos campos).
  """

  count_total = 0
  count_cold = 0
  for row in data:
    for posicao, col in enumerate(row):
      if posicao == get_nr_coluna(data, "type"):
        count_total += 1
        if row[posicao] == 'C':
          count_cold += 1
  razao_frios = count_cold/(count_total - 1)

  return 1-(razao_frios), razao_frios
```

Aplica a função count_ratio_hot_cold, que faz a contagem de valores Cold (C) e quantitativo total (Hot e Cold). Calcula a razão existente entre Cold/Registros totais, assim como Hot/Registros totais (excluindo a linha referente ao nome dos campos).

```python
cereais_quentes, cereais_frios = count_ratio_hot_cold(dados_novos)
```

evaluete_cereal filtra as linhas em que o rating é maior que 70 e retorna os nomes dos cereais.

```python
def evaluate_cereal(data):
  best_evaluated = []
  for row in data:
    for posicao, col in enumerate(row):
      if posicao == get_nr_coluna(data, "rating"):
        if (row[posicao] != 'rating') and (float(row[posicao]) > 70):
          cereal_position = get_nr_coluna(data, "name")
          cereal_name = row[cereal_position]
          best_evaluated.append(cereal_name)

  return best_evaluated
```

```python
melhores_cereais = evaluate_cereal(dados_novos)
```

A função a seguir responde a pergunta "Quais os cereais que possuem mais do que (incluído) 2 em fibra, possui vitamina e é servido quente?". No caso nenhum cereal atendia aos requisitos e o resultado foi uma lista vazia.

```python
def evaluate_fiber_vitamin_hot(data):

  match_condition = []
  for row in data:
    posicao_fiber = get_nr_coluna(data, "fiber")
    posicao_vitamina = get_nr_coluna(data, "vitamin")
    posicao_type = get_nr_coluna(data, "type")

    if (
        (row[posicao_fiber] != 'fiber')
        and (float(row[posicao_fiber]) >= 2)
        and (float(row[posicao_vitamina]) > 0)
        and (row[posicao_type] == "H")
    ):

      match_condition.append(row[get_nr_coluna(data, 'name')])

  return match_condition
```

```python
def get_column_as_list(data, nome):
  """Extrai uma dada coluna dos dados como lista"""

  lista_coluna = []

  for row in data:
    for posicao, col in enumerate(row):
      if posicao == get_nr_coluna(data, nome):
        if (row[posicao] != nome):
          lista_coluna.append(row[posicao])

  return lista_coluna
```

```python
lista_fibras = get_column_as_list(dados_novos, "fiber")
lista_carbo = get_column_as_list(dados_novos, "carbo")
lista_calories = get_column_as_list(dados_novos, "calories")
lista_protein = get_column_as_list(dados_novos, "protein")
lista_rating = get_column_as_list(dados_novos, "rating")
```

```python
def get_correlacao(lista_atributo1, lista_atributo2):
  """
  Realiza o cálculo da correlação de pearson entre duas listas numéricas de mesmo tamanho
  """
  numerador = 0
  denominador1 = 0
  denominador2 = 0

  lista_atributo1 = list(map(float, lista_atributo1))
  lista_atributo2 = list(map(float, lista_atributo2))
  media_atributo1 = sum(lista_atributo1) / len(lista_atributo1)
  media_atributo2 = sum(lista_atributo2) / len(lista_atributo2)

  for i in range(len(lista_atributo1)):
    numerador += ((lista_atributo1[i] - media_atributo1) * (lista_atributo2[i] - media_atributo2))
    denominador1 += (lista_atributo1[i] - media_atributo1)**2
    denominador2 += (lista_atributo2[i] - media_atributo2)**2

  return numerador / (denominador1 * denominador2)**0.5
```

```python
cor_fib_carbo = get_correlacao(lista_fibras, lista_carbo)
cor_cal_carbo = get_correlacao(lista_calories, lista_carbo)
cor_prot_rating = get_correlacao(lista_protein, lista_rating)
```

```python
print(cor_fib_carbo)
print(cor_cal_carbo)
print(cor_prot_rating)
```

Os resultados indicam uma correlação fraca negativa entre fibras e carboidratos, correlação desprezível entre quantidade de calorias e carboidratos e uma correlação fraca positiva entre proteínas e o rating.

A função sort_list_by_rating faz a conversão para float dos dados nessa coluna e ordena de forma crescente.

```python
from operator import itemgetter

def sort_list_by_rating(data):
  """Recebe os dados e retorna a lista de listas ordenada pela coluna de rating"""
  posicao_rating = get_nr_coluna(data, "rating")
  for row in data:
    for posicao, col in enumerate(row):
      if posicao == posicao_rating:
        if (row[posicao] != 'rating'):
          row[posicao] = float(row[posicao])

  get_n = itemgetter(posicao_rating)
  sorted_data = sorted(data[1:len(data)], key=get_n)
  new_data = [data[0]] + sorted_data
  return new_data
```

```python
print(sort_list_by_rating(dados_novos))
```
